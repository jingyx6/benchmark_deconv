## 4. (Required) Define Functions

```python
def simulate_reference_split(data_in, 
                             select_cell, select_tissue,
                             scenarios,
                             tissue_varname='tissue', ct_varname='cell_type', subject_varname='study'
                             ):
    df_meta = data_in[(data_in.obs[tissue_varname].isin(Tissue)) & (data_in.obs[ct_varname].isin(cells))]
    if scenarios == 'Same subject':
        study_id_sim = np.random.choice(df_meta.obs[subject_varname].unique(), size=1)
        study_id_deconv = study_id_sim
    elif scenarios == 'Different subject':
        study_id_sim, study_id_deconv = np.random.choice(adata.obs[subject_varname].unique(), size=2, replace=False)
        study_id_sim = [study_id_sim]
        study_id_deconv = [study_id_deconv]
    else:
        study_id_sim = np.random.choice(df_meta.obs[subject_varname].unique(), size=1)
        study_id_deconv = [st for st in df_meta.obs[subject_varname].unique() if st not in study_id_sim]
    print(f'scenarios: {scenarios}')
    print("for simulation: ", study_id_sim )
    print("for deconv: ", study_id_deconv )
    sim_ids = list(df_meta[df_meta.obs['study'].isin(study_id_sim)].obs_names)
    print("total scExp for simulation: ",len(sim_ids))
    ref_ids = list(df_meta[df_meta.obs['study'].isin(study_id_deconv)].obs_names)
    print("total scExp for deconv reference: ",len(ref_ids))
    sim_adata = df_meta[df_meta.obs_names[df_meta.obs_names.isin(sim_ids)], df_meta.var_names].copy()
    ref_adata = df_meta[df_meta.obs_names[df_meta.obs_names.isin(ref_ids)], df_meta.var_names].copy()
    sim_adata.uns['configure']={'Tissue': select_tissue,
                            'CancerousCells': CancerousCells,
                            'Scenarios': scenarios,
                            'Subject' : study_id_sim,                           
                            'CellTypeList': select_cell,
                            'CellTypeNumber': len(select_cell)}
    ref_adata.uns['configure']={'Tissue': select_tissue,
                            'CancerousCells': CancerousCells,
                            'Scenarios': scenarios,
                            'Subject' : study_id_deconv,                           
                            'CellTypeList': select_cell,
                            'CellTypeNumber': len(select_cell)}
    return sim_adata, ref_adata
```

```python
def generateMultiN(pattern_gp_label,pattern_main_dict, pattern_minor_dict, ipt, ntotal, mix, mix_level, ct_select):
    print("Generating cell type proportions for pattern", ipt)
    nobs = np.sum(pattern_gp_label == ipt) ## n spot where region label == ipt
    sample = np.zeros((nobs, len(ct_select))) ## a matrix nobs * ct_select cell types
    sample_names = np.array(list(pattern_gp_label.keys()))[pattern_gp_label == ipt] ## get spot ids
    prop = np.zeros((nobs, len(ct_select)))
    if pattern_minor_dict[ipt] == 'random': ### if monor set random, while random sample total number of cell types
        numCT = np.random.choice(np.arange(1, len(ct_select) + 1))  # sample total number of cell types in each layer
        pattern_minor_dict[ipt] = ['cell random']*(numCT-1)
    else:
        numCT = 1 + len(pattern_minor_dict[ipt])

    # cell_type_count_dict = dict((x,pattern_minor_dict[ipt].count(x)) for x in set(pattern_minor_dict[ipt])) ### count minor cell type 
    main, minor = [], []
    main = pattern_main_dict[ipt] # defined one dominant cell type in layer ipt
    minor = pattern_minor_dict[ipt] ## define minor cell ype in layer ipt
    concen = np.ones(numCT)  # set the concentration parameters, currently fixed to be 1 for all cell types

    propSample = np.random.dirichlet(concen, size=nobs)
    ct_select_sub_sample = np.random.choice(list(set(ct_select) - set(minor + [main])), size=minor.count('cell random'), replace=False)
    if len(ct_select_sub_sample)>0:
        ct_select_sub = [main] + list(set(ct_select) & set(minor)) + list(ct_select_sub_sample)
    else:
        ct_select_sub =  [main] + list(set(ct_select) & set(minor))
    print('ct_select_sub:test' + str(ct_select_sub))
    fix_dirichlet = np.random.choice(np.arange(nobs), size=int(nobs * (1-mix[mix_level])), replace=False) ## dominant spot 
    mix_dirichlet = np.setdiff1d(np.arange(nobs), fix_dirichlet) ## add noise mix spot, dominant cell type may change

    if len(mix_dirichlet) > 0:
        propSample[mix_dirichlet] = np.random.dirichlet(np.ones(numCT), size=len(mix_dirichlet))

    print(f'''-----configure-------\n
    main: {main} \nminor:  {minor}\n
    --- Final assign -----\n
    main: {main}\n
    cell combine: {ct_select_sub}\n
    total # {numCT} cell types assigned ''')
    if len(fix_dirichlet) > 0:
        for i in fix_dirichlet:
            propSample[i] = propSample[i][np.argsort(propSample[i])[::-1]]  # order the proportion to assign the largest proportion to the dominant cell type

    prop[:, [list(ct_select).index(x) for x in ct_select_sub]] = propSample # replace selected cell type with dominant proportion 
    sample = np.round(ntotal * prop).astype(int)

    # # To avoid no numbers sampled
    index = np.where(np.sum(sample, axis=1) == 0)[0]
    if len(index) > 0:
        print(f'row {index} have no cell type sampled, ramdom assign')
        for i in index:
            sample[i] = np.random.multinomial(ntotal, prop[i])   
    return {"sample": sample}
def pbmclapply_helper(isample, temp_nct, temp_exprs, ct_id, ct_select, sample_withRep=False):
    if isample % 50 == 0:
        print(f'Progress at {isample}')
    temp_sample = temp_nct.iloc[isample]
    temp_sample_pseudo = {}
    for ict in ct_select:
        temp_vec = temp_exprs[:, ct_id == ict]
        if temp_nct.iloc[isample][ict] > temp_vec.shape[1]:
            sample_withRep = True
        temp_id = np.random.choice(np.arange(temp_vec.shape[1]), size=int(temp_nct.iloc[isample][ict]), replace=sample_withRep) ### random sample sc gene expression of specific cell type 
        if len(temp_id) > 1:
            temp_sample_pseudo[ict] = pd.Series(np.array(temp_vec[:, temp_id].sum(axis=1)).ravel(), index=np.arange(temp_vec.shape[0])) ### if cell count > 1 for this cell type, sum gene count of those cells accordingly
        elif len(temp_id) == 1:
            temp_sample_pseudo[ict] = pd.Series(temp_vec[:, temp_id].toarray().reshape(-1), index=np.arange(temp_vec.shape[0])) ### if cell count = 1 for this cell type, then find gene count of one cell accordingly
        elif len(temp_id) == 0:
            temp_sample_pseudo[ict] = pd.Series(0, index=np.arange(temp_vec.shape[0])) ### if cell count = 0 for this cell type, gene count = 0 accordingly
    return pd.DataFrame(temp_sample_pseudo).sum(axis=1) ## sum gene expression in spot

def generateSpatial_norep_fixedProp(seed, eset_sub_split1, ct_varname, ct_select, 
                                    sample_withRep, pattern_gp_label, ntotal, mix, mix_level):
    time1 = time.time()
    # phenoData = eset_sub_split1.phenoData.data
    k = len(set(ct_select))
    print(f"Using {k} cell types to generate pseudo spatial dataset")

    Sample_random = np.zeros((len(pattern_gp_label), len(ct_select)))
    Sample_random = pd.DataFrame(Sample_random, index=pattern_gp_label.keys(), columns=ct_select)

    ct_id = eset_sub_split1.obs[ct_varname]

    for ipt in np.sort(pattern_gp_label.unique()):
        pattern = generateMultiN(pattern_gp_label,pattern_main_dict, pattern_minor_dict, ipt, ntotal, mix, mix_level, ct_select)
        Sample_random.loc[pattern_gp_label == ipt] = pattern['sample']
    print(f"Generating pseudo spatial dataset for {len(set(pattern_gp_label))} patterns")

    np.random.seed(seed)
    temp_exprs = eset_sub_split1.X.T
    temp_nct = Sample_random
    true_p = temp_nct.divide(temp_nct.sum(axis=1), axis=0)
    print(temp_nct.shape)



    pool = Pool(processes=50)  # Set the number of processes to use
    pbmclapply_partial = partial(pbmclapply_helper,temp_nct=temp_nct, temp_exprs=temp_exprs, ct_id=ct_id, ct_select=ct_select, sample_withRep=False)
    temp_pseudo = pool.map(pbmclapply_partial, range(temp_nct.shape[0]))
    pool.close()
    pool.join()
    time2 = time.time()
    print("time useï¼š", (time2-time1)//60)
    temp_pseudo = np.vstack(temp_pseudo).T
    temp_pseudo = pd.DataFrame(temp_pseudo, columns=Sample_random.index, index=eset_sub_split1.var_names)
    
    return {"pseudo.data": temp_pseudo, "true.p": true_p, "ntotal": ntotal, "Sample_random": Sample_random}
```

```python
## assign atlas reference data
adata = sc.read_h5ad("./HLA_qc_curated_v1.h5ad")
```

```python
## Set the parametet for simulate and reference data set split
## In Scenario 1, we only include manor cell types, and adjust the cell count per spot.
dataorigin = 'hlca'
CancerousCells = 'No'
Tissue = ['lung parenchyma']
cells = ['fibroblast of lung',
         'epithelial cell of lung',
         'alveolar capillary type 1 endothelial cell',
         'B cell'] # This is the list of all ell types in your synthetic data (both major and minor cell types)
simulate_reference_split_scenarios_list = ['Same subject', 'Different subject', 'Atlas']
simulate_reference_split_scenarios = 0  # 0: same subject
scenarios = simulate_reference_split_scenarios_list[simulate_reference_split_scenarios]
subject_varname = 'study'
tissue_varname = 'tissue'
ct_varname = 'cell_type'
```

## 6. Generate reference data

```python
### split, this is a random process, but you can repeat this to get a specific study
sim_adata, ref_adata = simulate_reference_split(adata,select_cell=cells,
                                               select_tissue=Tissue,
                                               scenarios=scenarios,
                                               )
sim_adata.uns['configure']['Dataorigin'] = dataorigin
ref_adata.uns['configure']['Dataorigin'] = dataorigin
```

```python
### store it to local, note this will create a folder called store_id (ref_id)
store_id = hashlib.md5(str(sim_adata.uns['configure']).encode()).hexdigest()
print(f'''----Simulate Refernece data configure----\n {sim_adata.uns['configure']}\n
----Deconv Refernece data configure----\n {ref_adata.uns['configure']}\n
----Start save-------\n
store_id {store_id}
        ''')
store_path = os.path.join(path, store_id)
if not os.path.exists(store_path):
    os.makedirs(store_path)
sim_adata.write_h5ad(
    os.path.join(store_path,f"Reference_for_simulate.h5ad"),
    compression=hdf5plugin.FILTERS["zstd"]
)
ref_adata.write_h5ad(
    os.path.join(store_path,f"Reference_for_deconv.h5ad"),
    compression=hdf5plugin.FILTERS["zstd"]
)
```

## 7. Generate spatial synthetic data

```python
### read spatial informantion
spatial = pd.read_csv('./Square_sps_location_seed1.csv')
spatial['x']=spatial['x']*100
spatial['y']=spatial['y']*100
coordinates = np.asarray([[spatial.iloc[i]['x'],spatial.iloc[i]['y']] for i in spatial.index]) ### get spatial coordinate inf.
```

#### <font color="red"> Your Input Needed here to adjust "n_total" (cell density)

```python
## Input ## read in simulated reference data
eset_sub_split1 = sc.read_h5ad(store_path + '/Reference_for_simulate.h5ad')

### domain cell type for each region group
pattern_main_dict = dict(zip(np.sort(spatial.group.unique()), ['fibroblast of lung', 'epithelial cell of lung',
       'alveolar capillary type 1 endothelial cell',
       'B cell']))
pattern_minor_dict = dict(zip(spatial.group.unique(), [[], 
                                                       [],
                                                       [],
                                                       []]))


DataQuality = 'Clean'
pattern_gp_label = spatial['group']
mix = [0, 0.2, 0.4, 0.6]
mix_level = 0   # 0,1,2,3 representing mix 0,0.2,0.4,0.6
seed = 3
ntotal = 10  # Adjust this ntotal to 5, 10, 15, 20, 25 for Scenario 1 (cell density : N cells per spot)
ct_select = eset_sub_split1.obs.cell_type.unique()
```

```python
# Generate spatial synthetic data
spatial_pseudo = generateSpatial_norep_fixedProp(
        seed=seed,
        eset_sub_split1=eset_sub_split1,
        ct_varname=ct_varname,
        # sample_varname=sample_varname,
        ct_select=ct_select,
        sample_withRep=False,
        pattern_gp_label=pattern_gp_label,
        ntotal=ntotal,
        mix=mix, 
        mix_level=mix_level
    )
    
```

```python
### store into anndata
sim_srt_adata = AnnData(spatial_pseudo['pseudo.data'].T, 
                obsm={"spatial": coordinates}, 
                dtype=np.int64)
sim_srt_adata.uns['ground_truth']={"cell_proportion":spatial_pseudo['true.p'], 'cell_count':spatial_pseudo['Sample_random']}
sim_srt_adata.uns['configure']  = eset_sub_split1.uns['configure']
sim_srt_adata.uns['configure']['DataQuality'] = DataQuality
sim_srt_adata.uns['configure']['DomainCellTypeForEachRegion'] = pattern_main_dict
sim_srt_adata.uns['configure']['CellNumberPerSpot'] = ntotal 
sim_srt_adata.uns['configure']
```

```python
### save spatial synthetic data to local, dataset_id will be called sim_id hereafter
dataset_id = hashlib.md5(str(sim_srt_adata.uns['configure']).encode()).hexdigest()
sim_srt_adata.write_h5ad(
    os.path.join(store_path, f"simulated_SRT_dataset_{dataset_id}.h5ad"),
    compression=hdf5plugin.FILTERS["zstd"]
)
```
