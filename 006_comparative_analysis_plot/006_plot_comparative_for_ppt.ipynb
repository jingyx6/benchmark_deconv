```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the CSV file into a pandas DataFrame
all_RMSE = pd.read_csv("all_RMSE.csv")

# Step 2: Group the data by each method
methods_columns = all_RMSE.columns[-12:-2]  # the last 12 columns are the methods, ignore SpatialDWLS since the accuracy was really bad
methods = methods_columns
ref_list = all_RMSE.ref_source.unique()
```

```python
df = all_RMSE
# Assuming you want to drop the row with index 13 from DataFrame 'df'
df.drop(13, inplace=True) # this dataset has weird output due to a wrong setting when generating the synthetic data. remove it here 
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a DataFrame called 'df'
# and you have defined 'methods_columns' and 'palette' lists

# Set the number of subplots in a row
num_cols = 5

# Calculate the number of rows required to accommodate all subplots
num_rows = (len(methods_columns) - 1) // num_cols + 1

# Create the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows))

# Loop through each method and create a bar plot in the corresponding subplot
for i, method in enumerate(methods_columns):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]
    
    # Create the bar plot
    sns.barplot(x="ref_source", y=method, data=df, ci=95, palette='Set2', alpha =0.8, ax=ax)
    ax.set_xlabel("ref_source")
    ax.set_ylabel(method)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    
        
    # Overlay individual data points using swarmplot with different colors
    sns.swarmplot(x="ref_source", y=method, data=df, hue="cell_mix", palette={"NONE": "black", "Mixed": "blue"}, ax=ax)
    
# Remove any empty subplots
if len(methods_columns) < num_rows * num_cols:
    for i in range(len(methods_columns), num_rows * num_cols):
        fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```

### 1. Get the sorted method

```python
# Group the DataFrame by 'ref_source'
grouped_data = selected_df.groupby('ref_source')

# Define a function to sort and print the methods by mean RMSE for each 'ref_source'
def print_sorted_methods_by_ref_source(ref_source, ref_source_group):
    # Calculate the mean RMSE for each method in the 'ref_source' group
    method_means = ref_source_group[methods].mean()
    
    # Sort the methods based on their mean RMSE values in ascending order
    sorted_methods = method_means.sort_values()
    
    # Print the list of first 6 methods sorted by mean RMSE for the 'ref_source' group
    print(f"Methods sorted by Mean RMSE for ref_source: {ref_source}")
    print(' > '.join(sorted_methods.index.tolist()[:4]))

# Loop through each 'ref_source' group and print the sorted methods
for ref_source, ref_source_group in grouped_data:
    print_sorted_methods_by_ref_source(ref_source, ref_source_group)
```

### 2. Plot Scenario 2_1

```python
# Set the global font size to 15
plt.rcParams.update({'font.size': 15})

# Set the number of subplots in a row
num_cols = 3

# Define the desired order for the panels
desired_order = ['Same study','Different Study',  'Atlas']

# Calculate the number of rows required to accommodate all subplots
num_rows = (len(methods) - 1) // num_cols + 1

# Create the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows))

# Group the DataFrame by 'ref_source'
grouped_data = selected_df.groupby('ref_source')
# Define a colormap for the bars
colormap = plt.cm.Set3

# Loop through each group and create a subplot
for i, ref_source in enumerate(desired_order):
    group = grouped_data.get_group(ref_source)
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]
    
    # Plot the data for each 'ref_source'
    for j, method in enumerate(methods):
        # Plot the bar for the mean value
        ax.bar(method, group[method].mean(), yerr=group[method].std(), capsize=5, color=colormap(j), alpha=0.8)
        
        # Add individual data points as black dots on top of the bars
        x_positions = [j] * len(group[method])
        ax.scatter(x_positions, group[method], color='black', s=30, alpha=0.7)
       
    ax.set_xlabel("Methods")
    ax.set_ylabel("Mean RMSE")
    ax.set_title(f"ref_source: {ref_source}")
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_xticklabels(methods, rotation=45, ha="right")
    ax.set_ylim (0,0.45)
    
# Remove any empty subplots
if len(methods) < num_rows * num_cols:
    for i in range(len(methods), num_rows * num_cols):
        fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```

# Analyze RMSE for four regions in Scenario 3

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the CSV file into a pandas DataFrame
region_RMSE = pd.read_csv("/home/oneai/oneai-dda-spatialone-deconvbenchmark/EnDeconWorkflow/006_comparative_analysis_plot/Scenario3/senario3_starmap_groups_rmse.csv")

# Step 2: Group the data by each method
Methods = region_RMSE.columns[-11:-1]  # Assuming the last 10 columns are the methods
region_RMSE.rename(columns={region_RMSE.columns[2]: 'region'}, inplace=True)
region_list = region_RMSE.region.unique()

# Step 3: Find the minimum and maximum values in the entire DataFrame
min_value = region_RMSE[Methods].min().min()
max_value = region_RMSE[Methods].max().max()

# Step 4: Modify the 'dataset' column to keep only the 2nd string when split by '_'
region_RMSE['dataset'] = region_RMSE['dataset'].apply(lambda x: x.split('_')[1])

# Step 5: Plot heatmap for each method with a uniform color scale
fig, axes = plt.subplots(nrows=len(Methods)//5 + 1, ncols=5, figsize=(20, 4 * (len(Methods)//5 + 1)), sharex=True)

for idx, method in enumerate(Methods):
    row = idx // 5
    col = idx % 5
    
    # Filter data for the specific method
    method_data = region_RMSE.pivot(index='dataset', columns='region', values=method)
    
    # Create the heatmap plot for the current method with uniform color scale
    sns.heatmap(method_data, ax=axes[row, col], cmap='viridis', annot=True, fmt=".2f", cbar=False, vmin=0, vmax=0.1)
    
    # Set the title for each subplot
    axes[row, col].set_title(f'{method}')
    
    # Set the x and y axis labels
    axes[row, col].set_xlabel('Region')
    axes[row, col].set_ylabel('Dataset')

# Remove empty subplots
for idx in range(len(Methods), (len(Methods)//5 + 1) * 5):
    fig.delaxes(axes.flatten()[idx])

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
